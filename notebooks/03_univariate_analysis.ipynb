```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
enrolment_df = pd.read_csv("../data/clean_enrolment.csv")
biometric_df = pd.read_csv("../data/clean_biometric.csv")
demographic_df = pd.read_csv("../data/clean_demographic.csv")
```

```python
enrolment_df["date"] = pd.to_datetime(enrolment_df["date"])
biometric_df["date"] = pd.to_datetime(biometric_df["date"])
demographic_df["date"] = pd.to_datetime(demographic_df["date"])
```

```python
daily_enrolment = enrolment_df.groupby("date")["total_enrolment"].sum()

plt.figure(figsize=(10, 5))
daily_enrolment.plot()
plt.title("Total Aadhaar Enrolment Over Time")
plt.xlabel("Date")
plt.ylabel("Total Enrolments")
plt.show()
```

```python
age_totals = enrolment_df[["age_0_5", "age_5_17", "age_18_greater"]].sum()

plt.figure(figsize=(6, 6))
age_totals.plot(kind="pie", autopct="%1.1f%%")
plt.title("Age-wise Distribution of Aadhaar Enrolments")
plt.ylabel("")
plt.show()
```

```python
state_enrolment = (
    enrolment_df.groupby("state")["total_enrolment"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10, 5))
state_enrolment.plot(kind="bar")
plt.title("Top 10 States by Total Aadhaar Enrolment")
plt.xlabel("State")
plt.ylabel("Total Enrolments")
plt.show()
```

```python
bio_time = biometric_df.groupby("date")["total_biometric_activity"].sum()

plt.figure(figsize=(10, 5))
bio_time.plot()
plt.title("Total Biometric Activity Over Time")
plt.xlabel("Date")
plt.ylabel("Biometric Events")
plt.show()
```

```python
demo_time = demographic_df.groupby("date")["total_demographic_updates"].sum()

plt.figure(figsize=(10, 5))
demo_time.plot()
plt.title("Total Demographic Updates Over Time")
plt.xlabel("Date")
plt.ylabel("Demographic Updates")
plt.show()
```
